```python
import seaborn as sns
import matplotlib.pyplot as plt

big_tikets = ['Stored Value Card', 'Five Day Pass', 'Group Ticket']
counts = [150, 80, 120]


import pandas as pd
df = pd.DataFrame({'Ticket Type': big_tikets, 'Count': counts})


sns.barplot(x='Ticket Type', y='Count', data=df)

plt.xticks(rotation=100)
plt.title('Количество купленных билетов за все время')
plt.show()
```
